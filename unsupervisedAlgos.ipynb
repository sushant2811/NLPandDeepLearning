### Non-negative matrix factorization (NMF)

```python
from sklearn.decomposition import NMF
```

```python
nmf = NMF(n_components=6)
```

```python
import pandas as pd
```

### Images

```python
images = pd.read_csv('./datasetsUnsupervised/lcd-digits.csv', header = None)
```

```python
images.head()
```

```python
imagesArr = images.as_matrix()
```

```python
imagesArr.shape
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
digit = imagesArr[0, :]
bitmap = digit.reshape(13, 8)
plt.imshow(bitmap, cmap = 'gray', interpolation='nearest');
plt.colorbar();
```

```python
nmfImages = NMF(n_components=7)
```

We use 7 components above, because 7 is the number of cells in the LED display

```python
def show_as_image(sample):
    bitmap = sample.reshape((13, 8))
    plt.figure()
    plt.imshow(bitmap, cmap='gray', interpolation='nearest')
    plt.colorbar()
    plt.show()
```

```python
imagesArr.shape
```

```python
featuresIm = nmfImages.fit_transform(imagesArr)
```

```python
featuresIm
```

```python
featuresIm.shape
```

```python
nmfImages.components_.shape
```

```python
for component in nmfImages.components_:
    
    show_as_image(component)
```

```python
featuresIm[0]
```

Note how the feature for the first number has non-zero entries such that it builds up to number 7. 

**Comparison with PCA**

```python
from sklearn.decomposition import PCA
```

```python
pcaImages = PCA(n_components=7)
```

```python
featuresImPCA = pcaImages.fit_transform(imagesArr)
```

```python
for component in pcaImages.components_:
    
    show_as_image(component)
```

Unlike NMF, PCA doesn't learn the parts of things. Its components do not correspond to topics (in the case of documents) or to parts of images, when trained on images. 

**Recommender system using NMF**
