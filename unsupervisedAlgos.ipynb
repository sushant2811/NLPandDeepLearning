### Non-negative matrix factorization (NMF)

```python
from sklearn.decomposition import NMF
```

```python
nmf = NMF(n_components=6)
```

```python
import pandas as pd
```

```python
from scipy.sparse import csr_matrix
```

```python
df = pd.read_csv('./datasetsUnsupervised/Wikipedia articles/wikipedia-vectors.csv', index_col=0)
```

```python
df.shape
```

```python
df.head()
```

```python
articles = csr_matrix(df.transpose())
```

```python
df.transpose()
```

```python
articles.shape
```

```python
articles
```

```python
titles = list(df.columns)
```

```python
titles
```

```python
len(titles)
```

```python
nmf.fit(articles)
```

```python
nmfFeatures = nmf.transform(articles)
```

```python
nmfFeatures
```

```python
nmfFeatures.shape
```

Now, let's make sense of the features

```python
dfFeatures = pd.DataFrame(nmfFeatures, index = titles)
```

```python
dfFeatures
```

```python
dfFeatures.loc['Anne Hathaway']
```

```python
dfFeatures.loc['Denzel Washington']
```

```python
nmf.components_
```

```python
nmf.components_.shape
```

```python
dfFeatures.shape
```

```python
words = pd.read_csv('./datasetsUnsupervised/Wikipedia articles/wikipedia-vocabulary-utf8.txt', header=None)
```

```python
words.dtypes
```

```python
wordsL = list(words.values[:,0])
```

```python
components_df = pd.DataFrame(nmf.components_, columns=wordsL)
```

```python
nmf.components_
```

```python
components_df
```

```python
component3 = components_df.iloc[3]
```

```python
component3.nlargest(30)
```

```python
components_df.iloc[1].nlargest(30)
```

```python
components_df.iloc[2].nlargest(30)
```

```python
components_df.iloc[4].nlargest(30)
```

Note how different types of words have the highest value in each components.
